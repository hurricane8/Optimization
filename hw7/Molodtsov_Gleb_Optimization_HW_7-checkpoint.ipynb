## Задача 1

Рассмотрим "распределенную" задачу минимизации:
\begin{equation}
\min_{x \in \mathbb{R}^d} f(x) = \frac{1}{n} \sum\limits_{i=1}^n f_i(x) = \frac{1}{2n} \sum\limits_{i=1}^n x^T A_i x - b^T_i x,
\end{equation}
где $n = 10$, в качестве матриц $\{A_i\}_{i=1}^n$ - случайные матрицы с $L =1000$, $\mu = 1$, $d = 100$, $\{b_i\}_{i=1}^n$ - так же случайные векторы.

__(а)__ Для такой задачи симулируем распределенное решение (вычисления на 1 компьютере, без распараллеливания). 
Реализуйте обычный распределенный градиентый спуск с операторами сжатия. Сжимайте информацию только от устройств к серверу. Используйте операторы сжатия $\text{Rand}1\%$, $\text{Rand}5\%$, $\text{Rand}10\%$, $\text{Rand}20\%$. Для каждого случая подберите шаг $\gamma$ для наилучшей сходимости, постройте сравнительный график сходимости (по оси $y$ - критерий $\| \nabla f(x^k)\|^2$, а по оси $x$ - число "переданной информации" от устройств к серверу). Сделайте вывод.

```python
import numpy as np
from math import log, exp, ceil, sqrt
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib import ticker
from sklearn.datasets import load_svmlight_file
from sklearn.linear_model import LinearRegression
import time 
from scipy.stats import special_ortho_group  
import matplotlib.gridspec as gridspec
import random
```

```python
## Adding constants
L = 1000
mu = 1
d = 100
E = np.eye(d)
n = 10
```

```python



def generate_matrix(d_local, mu_local, L_local):
    i = 0
    while i == 0:
        diagonal_entries = np.random.uniform(mu_local, L_local, d_local)
        diagonal_entries[0] = mu_local
        diagonal_entries[1] = L_local
        D = np.diag(diagonal_entries)
        m = special_ortho_group.rvs(d_local)
        A = (np.linalg.inv(m).dot(D)).dot(m)
        np.set_printoptions(suppress=True)
        if (is_pos_def(A-mu_local*E)):
            i += 1
    return A
    
def is_pos_def(x): #проверка положительной определённости
    return np.all(np.linalg.eigvals(x) > 0)

def gradient(x): 
    grad_ = np.zeros(d)
    for j in range(n):
        grad_ += A[j] @ x - b[j]
    return grad_/n
```

```python
start = np.zeros(d)
A = []
b = []
for i in range (n):
    A.append(generate_matrix(d, mu, L))
    b.append(np.random.uniform(0, 10, d))
A = np.array(A)
b = np.array(b)
print(A.shape)
print(b.shape)
```

```python
def RAND(x, rand_):
    grad_ = np.zeros(d)
    num = np.random.randint(0, d, rand_)
    for j in range(n):
        for i in (num):
            grad_[i] += A[j][i]@(x) - b[j][i]
    #print(grad_)
    return grad_

def gradient_descent_rand(start, learn_rate, n_iter, rand):
    values_x = []
    values_y = []
    vector = start
    vector_list = []
    begin_time = time.time()
    timeprog = []
    operation_number = 0
    for i in range(n_iter):
        grad = RAND(vector, rand) / n
        operation_number += n * rand
        values_x.append(operation_number)
        gradientt = gradient(vector)
        values_y.append(gradientt.T.dot(gradientt))
        vector = vector - learn_rate * (grad)
        vector_list.append(vector)
        timeprog.append(time.time() - begin_time)
    return values_x, values_y, vector_list, timeprog
```

```python
def TOP(grad_, top_):
    num = np.argsort(-np.absolute(grad_))[:top_]
    #print(len(num))
    for i in range(d):
        if i not in num:
            grad_[i] = 0
    return grad_



def gradient_descent_top_K(start, learn_rate, n_iter, K):
    values_x = []
    values_y = []
    vector = start
    vector_list = []
    begin_time = time.time()
    timeprog = []
    for i in range(n_iter):
        gradientt = gradient(vector)
        grad = TOP(gradientt, K)/n
        values_x.append(i + 1)
        values_y.append(gradientt.T.dot(gradientt))
        vector_list.append(vector)
        vector = vector - learn_rate * (grad)
        timeprog.append(time.time() - begin_time)
    return values_x, values_y, vector_list, timeprog
```
