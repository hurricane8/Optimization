## Задача 1

Рассмотрим задачу минимизации:
\begin{equation}
\min_{x \in \mathbb{R}^d} f(x) = \frac{1}{2}x^T A x - b^T x,
\end{equation}
где в качестве матрицы $A$ - случайная матрица с $L =1000$, $\mu = 1$, $d = 100$, $b$ - так же случайный вектор.

__(а)__ Для такой задачи симулируем итерацию координатного спуска (смотри семинар). 

Для такого метода подберите шаг $\gamma$ для наилучшей сходимости постройте график сходимости (по оси $y$ - критерий $\| \nabla f(x^k)\|^2$, а по оси $x$ - число арифмитических операций при подсчете "градиента"). Постройте на этом же графике сходимость честного градиентного спуска c наилучшим шагом сходимости. Сделайте вывод о характере сходимости нового метода.

```python
import numpy as np
from math import log, exp, ceil, sqrt
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib import ticker
from sklearn.datasets import load_svmlight_file
from sklearn.linear_model import LinearRegression
import time 
from scipy.stats import special_ortho_group  
import matplotlib.gridspec as gridspec
import random
```

```python
## Adding constants
L = 1000
mu = 1
gamma = 1/L
d = 100
E = np.eye(d)
```

```python
def generate_matrix(d_local, mu_local, L_local):
    i = 0
    while i == 0:
        diagonal_entries = np.random.uniform(mu_local, L_local, d_local)
        diagonal_entries[0] = mu_local
        diagonal_entries[1] = L_local
        D = np.diag(diagonal_entries)
        m = special_ortho_group.rvs(d_local)
        A = (np.linalg.inv(m).dot(D)).dot(m)
        np.set_printoptions(suppress=True)
        if (is_pos_def(A-mu_local*E)):
            i += 1
    return A
    
def is_pos_def(x): #проверка положительной определённости
    return np.all(np.linalg.eigvals(x) > 0)

def gradient(x): 
    return A.dot(x) - b
```

```python
start = np.ones(d)/d
A = generate_matrix(d, mu, L)
b = np.random.uniform(0, 10, d)
```

```python
def stokh_gradient(x, i):
    res = np.zeros(d)
    res[i] = (A[i].dot(x) - b[i])
    return (res)

def gradient_descent(start, learn_rate, n_iter):
    values_x = []
    values_y = []
    vector = start
    vector_list = []
    begin_time = time.time()
    timeprog = []
    operation_number = 0
    for i in range(n_iter):
        grad = gradient(vector)
        operation_number += 2*d**2
        values_x.append(operation_number)
        values_y.append(grad.T.dot(grad))
        vector = vector - learn_rate * (grad)
        vector_list.append(vector)
        timeprog.append(time.time() - begin_time)
    return values_x, values_y, vector_list, timeprog

def gradient_descent_coord(start, learn_rate, n_iter):
    values_x = []
    vector = start
    vector_list = []
    begin_time = time.time()
    timeprog = []
    operation_number = 0
    for i in range(n_iter):
        j = random.randrange(d)
        stokh_grad = stokh_gradient(vector, j)
        operation_number += (2*d)
        values_x.append(operation_number)
        vector = vector - learn_rate * (stokh_grad)
        vector_list.append(vector)
        timeprog.append(time.time() - begin_time)
    return values_x, vector_list, timeprog
```

## Задача 2

Рассмотрим задачу линейной регрессии на датасете mushrooms с $\ell_2$-регуляризацией (коэффициент регуляризации равен $\frac{L}{1000}$). 

__(а)__ Оцените $\mu$ и $L$ для данной задачи. 

Найдём $L$: Для подстановки в условие липшецивости градиента, посчитаем сам градиент:

$\nabla(f(\omega)) = \nabla(\|X\omega - y\|^2 + \lambda\|\omega\|^2).\\$
Из предыдущих дз:
$\nabla(f(\omega)) = 2X^Т(X\omega - y) + \lambda\omega.$

Липшицев градиент с константой L: $\|(2X^ТX + \lambda I)(\omega_1 - \omega_0)\|\leq\|2X^ТX + \lambda I\|\cdot\|\omega_1 - \omega_0\|\Rightarrow L\geq\|2X^ТX + \lambda I\|$

Таким образом: $L$ - максимальное собственное значение матрицы $\frac{1}{n}(2X^ТX + \lambda I)$
Вспомним также, чему равно $\nabla^2(f(\omega))$. $\nabla^2(f(\omega)) = 2X^ТX + \lambda I$

Тогда $\mu$ - минимальное собственное значение матрицы $\frac{1}{n} (2X^ТX + \lambda I)$

__(б)__ Реализуйте координатный SGD, SEGA для задачи линейно регрессии. Каким будете брать $b$? Как будете выбирать шаг и другие параметры методов?

Величину b нужно выбирать так, чтобы она была сильно меньше, чем полная размерность (d). Тогда можно будет говорить, что одна итерация требует $O(1)$ подсчетов компонент градиента. Если же выбрать b слишком большой, но это уже будет не модификация SGD, а скорее модификация GD с неполным подсчетом градиента. Шаг выберем $\gamma ^* = \frac{0.8}{Ld}$ - для него получилась хорошая сходимость

```python
dataset = "mushrooms.txt" 

data = load_svmlight_file(dataset)
X_all, y_all = data[0].toarray(), data[1]
n_all, d_all = X_all.shape
#print(y)
for i in range (len(y_all)):
    if (y_all[i] == 2):
        y_all[i] = -1
#print(y)
X, X_test, y, y_test = train_test_split(X_all, y_all, train_size=0.81, random_state=42)

#####
n, d = X.shape

#####
L = np.linalg.eigvals(1/(2*n)*X.T.dot(X))[0]
x_0 = np.linalg.pinv(X) @ y



L = 1/n * max(np.linalg.eigvalsh((2 * X.T@X + L/1000 * np.identity(d))))
print(f"L = {L} \n")
mu = 1/n  * min(np.linalg.eigvalsh((2 * X.T@X + L/1000 * np.identity(d))))
print(f"mu = {mu} \n")
llambda = L/1000
print(f"lambda = {llambda} \n")
```

```python
def gradient_linreg_regul(omega): 
    return 1 / (2 * n) * X.T.dot(X.dot(omega) - y) + llambda * omega

def coord_gradient_batch_linreg_regul(omega, batch):
    res = np.zeros(d)
    for i in batch:
        res[i] = X.T[i]@(X@omega - y) + llambda * omega[i] ##еще 2 операции
    #print(res)
    return (1 / (2 * n) * res) ## итого 3 + b * (2 + n * (2d+1)) операций
```

```python
def gradient_descent_coord_several_dependent_linreg_regul(start, learn_rate, n_iter, b1):
    values_x = []
    vector = start
    vector_list = []
    begin_time = time.time()
    timeprog = []
    operation_number = 0
    schet = 0
    batches = create_batches(b1)
    for i in range(n_iter):
        j = batches[schet]
        stokh_grad = coord_gradient_batch_linreg_regul(vector, j)
        operation_number += 3 + b1 * (2 + n * (2 * d + 1))
        values_x.append(operation_number)
        vector = vector - learn_rate * (stokh_grad)
        vector_list.append(vector)
        timeprog.append(time.time() - begin_time)
        if (schet == len(batches) - 1):
            batches = create_batches(b1)
            schet = 0
        else:
            schet += 1
        #print(schet)
    return values_x, vector_list, timeprog
```
